# Tiền xử lý dữ liệu

```python
#Xử lý dữ liệu bị thiếu
dataset = dataset.dropna()
dataset.info()
```

```python
#Tạo một thuộc tính TotalCost = Quantity x UnitPrice.
dataset["TotalCost"] = dataset["Quantity"]*dataset["UnitPrice"]
dataset.head()
```
